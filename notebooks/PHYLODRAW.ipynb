```python
from Bio import Phylo
```

```python
tree = Phylo.read("nextstrain_ncov_gisaid_global_timetree.nwk", "newick")
```

```python
#print a representation of the tree:
print(tree) 
```

```python
# From this we see the datastructure.  It gives branch lengths,
# tree geometry (all those parentheses), and leaf names.
# I don't see anything else, though we might be able to get some
# meaningful fields from parsing the leaf names.

# hCoV-19/Wuhan/Hu-1/2019 
# hCoV-19/Colombia/VAC-ICE-SCOV-092/2021
```

```python
Phylo.draw_ascii(tree)
```

```python
Phylo.draw(tree)
```
